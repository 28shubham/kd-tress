```python
import pandas as pd
data= pd.read_csv("Shubham Thawait - Dataset.csv")
data = data.sort_values(by=[*data.columns[1:]])
data = data.values.tolist()
#print(data)
class Node:
    def __init__(self,point,left=None,right=None):
        self.point = point
        self.left = left
        self.right = right

def kdtree(tree,depth=0) :
  if not tree:
    return None


  axis = depth % 4

  sort = sorted(tree , key=lambda point: point[axis])
  medi = len(sort)//2
  medp = sort[medi]

  lsubtree = kdtree(sort[:medi],depth+1)
  rsubtree =kdtree(sort[medi+1:],depth+1)
  return Node(medp,lsubtree,rsubtree)

def searchkdtree(root, target, depth=0):
    if root is None:
        return None


    axis = depth % 4

    if target == root.point:
        return root.point

    if target[axis] < root.point[axis]:
        return searchkdtree(root.left, target, depth + 1)
    else:
        return searchkdtree(root.right, target, depth + 1)

def delete_node(root, point, depth=0):

    if root is None:
        return None


    axis = depth % 4

    if root.point == point:
        if root.right:
            min_node = min(root.right, axis, depth + 1)
            root.point = min_node.point
            root.right = delete_node(root.right, min_node.point, depth + 1)
        elif root.left:
            min_node = min(root.left, axis, depth + 1)
            root.point = min_node.point
            root.right = root.left
            root.left = None
        else:
            return None
    elif point[axis] < root.point[axis]:
        root.left = delete_node(root.left, point, depth + 1)
    else:
        root.right = delete_node(root.right, point, depth + 1)

    return root

def min(node, axis, depth):

    if node is None:
        return None


    naxis = depth % 4

    if naxis == axis:
        if node.left:
            return min(node.left, axis, depth + 1)
        return node
    else:
        left_min = min(node.left, axis, depth + 1)
        right_min = min(node.right, axis, depth + 1)
        min_node = node
        if left_min and left_min.point[axis] < min_node.point[axis]:
            min_node = left_min
        if right_min and right_min.point[axis] < min_node.point[axis]:
            min_node = right_min

        return min_node

tree = data
node = kdtree(tree)



deletenode = (6, 12,7,8)
node = delete_node(node, deletenode)
output = searchkdtree(node, deletenode)
if output:
    print(f" present {deletenode} .")
else:
    print(f" deleted {deletenode} .")
```
